GridSearchCV

For each classifier, I will define the parameter grid. For SVM different kernels. For Logistic Regression and KNN, will vary regularization and the number of neighbors.


```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
```

```python
# Define the hyperparameter grid for support vector machine     

Cs = [0.001, 0.01, 0.1, 1, 2, 5, 10]
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
```

```python
# Parameters for grid search
svc_class = SVC(kernel='linear')
svm_params = {"C": Cs, "kernel": kernels}
Cs, kernels
```

```python
svm_grid = GridSearchCV(svc_class, param_grid= svm_params, cv=5, verbose=2)
```

```python
svm_grid.fit(X_train_scaled,y_train)
```

```python
# Get the best hyperparameters
svm_grid.best_params_
```

```python
# Get the best score
svm_grid.best_score_
```

```python
# Get the best estimator and evaluate it on the test set
svm_grid.best_estimator_.score(X_test_scaled, y_test)
```

```python
from sklearn.metrics import classification_report

# Use the best estimator from the grid search to make predictions on the test set
svm_best = svm_grid.best_estimator_
```

```python
# Make predictions on the test set
y_pred_svm = svm_best.predict(X_test)
```

```python
# Fit PCA on scaled data and transform it
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_train_scaled)
```

```python
# Get the absolute values of the contribution of each feature
components_abs = np.abs(pca.components_)
```

```python
# Get the feature names
feature_names = list(X.columns)
```

```python
from sklearn.linear_model import LogisticRegression

# Fit a logistic regression model on the PCA transformed data
model = LogisticRegression()
# Fit the model
model.fit(X_pca, y_train)
```

```python
# Plotting the feature importances for the first two principal components
plt.figure(figsize=(12, 6))

# Plot for the first principal component
plt.subplot(1, 2, 1)
plt.barh(feature_names, components_abs[0])
plt.title('Principal Component 1')

# Plot for the second principal component
plt.subplot(1, 2, 2)
plt.barh(feature_names, components_abs[1])
plt.title('Principal Component 2')

# Display the plots tight layout and show
plt.tight_layout()
plt.show()
```

```python
# For each principal component, identify the feature with the highest contribution
top_features = [feature_names[components_abs[i].argmax()] for i in range(components_abs.shape[0])]
```

```python
# Create labels for the plot based on the top features
labels = [f"PC1: {top_features[0]}", f"PC2: {top_features[1]}"]
```

```python
# Create a DataFrame for the PCA results
pca_df = pd.DataFrame(data=X_pca, columns=labels)
pca_df['DrivingStyle'] = y_train.values
```

```python
# Assuming pca_df['DrivingStyle'] currently contains numeric labels
pca_df['DrivingStyle'] = pca_df['DrivingStyle'].map({0: 'AggressiveStyle', 1: 'EvenPaceStyle'})
```

```python
# Assuming the labels for PCA components are stored in `labels`
plt.figure(figsize=(10, 8))
sns.scatterplot(x=labels[0], y=labels[1], hue='DrivingStyle', data=pca_df, palette='viridis', alpha=0.7)
plt.title('PCA of Driving Styles with Influential Features')
plt.xlabel(labels[0]) # Set the x-axis label
plt.ylabel(labels[1]) # Set the y-axis label
plt.legend(title='Driving Style', loc='best')
plt.show()
```

Usine SMOTE to balance the data using x and y before PCA. SMOTE to oversample the minority class in the training data

```python
from imblearn.over_sampling import SMOTE

# Initialize SMOTE
smote = SMOTE(random_state=377123)

# Fit SMOTE to the training data and oversample the minority class to achieve a balanced class distribution
X_train_smote, y_train_smote = smote.fit_resample(X_train_scaled, y_train)

# Check the class distribution after applying SMOTE
from collections import Counter
print(f"Original class distribution: {Counter(y_train)}")
print(f"Resampled class distribution: {Counter(y_train_smote)}")
```
